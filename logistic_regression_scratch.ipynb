```python
import numpy as np
import matplotlib.pyplot as plt

# Define sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define cost function
def cost_function(X, y, theta):
    z = X.dot(theta) 
    h = sigmoid(z)
    cost = (-y * np.log(h) + (1 - y) * np.log(1 - h)).mean()
    #computes the derivative of the cost function with respect to each parameter
    grad = X.T.dot(h - y) / len(y)
    return cost, grad

# Define gradient descent function
def gradient_descent(X, y, theta, learning_rate, num_iterations):
    cost_history = []
    for i in range(num_iterations):
        cost, grad = cost_function(X, y, theta)
        theta -= learning_rate * grad
        cost_history.append(cost)
    return theta, cost_history

# Load data




X = np.array(data['x']).reshape(-1, 1)
y = np.array(data['y'])

# Feature scaling
X = (X - np.mean(X)) / np.std(X)

# Add bias term to X
X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)

# Initialize parameters
theta = np.zeros(X.shape[1])

# Run gradient descent
theta, cost_history = gradient_descent(X, y, theta, learning_rate=0.1, num_iterations=1000)

# Plot sigmoid function
x_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_plot = sigmoid(theta[0] + theta[1] * x_plot)
plt.plot(x_plot, y_plot)

# Plot data
plt.scatter(X[:, 1], y)

# Plot cost over iterations
plt.figure()
plt.plot(np.arange(len(cost_history)), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost over Iterations')

plt.show()
```
